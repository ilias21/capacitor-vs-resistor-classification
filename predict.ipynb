```python
from tensorflow.keras.models import load_model

import tensorflow as tf 
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**1. charger le model**

```python
model = load_model("./models/cap_vs_res.h5")
model_reg = load_model("./models/cap_vs_res_with_reg.h5")
```

**2. charger et preprocesser l'image**

```python
image = cv2.imread("./test.jpeg")
plt.imshow(image)
plt.show()

resized_image = tf.image.resize(image,(256,256))
plt.imshow(resized_image.numpy().astype(int))
plt.show()

resized_image = np.expand_dims(resized_image.numpy()/255,axis=0)
print(resized_image.shape)
predict = model.predict(resized_image)
if predict > 0.5:
    print("l'image reprsente une resistance")
else:
    print("l'image reprsente un capaciteur")
```
